```python
# importing modules

import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
from PIL import Image
```

```python
image = cv.imread('pvmodule.jpg')
# create a CLAHE object (Arguments are optional).

# to create histogram equalization
gray = cv.cvtColor(image, cv.COLOR_BGR2GRAY)
clahe = cv.createCLAHE(clipLimit=0.001, tileGridSize=(8,8))
cl1 = clahe.apply(gray)
plt.imshow(cl1, cmap='gray')
```

```python
blur = cv.GaussianBlur(gray, (5, 5), 0)
plt.imshow(blur, cmap='gray')
```

```python
canny = cv.Canny(blur, 30, 150, 3)
plt.imshow(canny, cmap='gray')
```

```python
dilated = cv.dilate(canny, (10, 10), iterations=1)
plt.imshow(dilated, cmap='gray')
```

```python
np.count_nonzero(dilated == 1)
```

```python
cv.imwrite("output.jpg", dilated)
```

```python
(cnt, hierarchy) = cv.findContours(
	canny.copy(), cv.RETR_CCOMP, cv.CHAIN_APPROX_NONE)
rgb = cv.cvtColor(image, cv.COLOR_BGR2RGB)
cv.drawContours(rgb, cnt, -1, (0, 255, 0), 2)

plt.imshow(rgb)
cv.imwrite("out2.jpg", rgb)
```
